```python
import pandas as pd
df = pd.read_csv("resultados_experimentos.csv")
df.head()
```

```python
def clasificar_biomasa(b):
    if b > 250:
        return "alta"
    elif 150 <= b <= 250:
        return "media"
    else:
        return "baja"

# Crear nueva columna con la clasificación
df['nivel_biomasa'] = df['biomasa'].apply(clasificar_biomasa)

# Verificar resultado
df[['sitio','biomasa','nivel_biomasa']].head()
```

```python
# Agrupar por nivel de biomasa y calcular promedios
resumen = df.groupby('nivel_biomasa')[['temperatura','humedad','luz']].mean()

print("Tabla de promedios agrupados:")
print(resumen)
```

```python
conteo = df['nivel_biomasa'].value_counts()

print("Número de sitios por nivel de biomasa:")
print(conteo)
```

```python
import matplotlib.pyplot as plt

# Gráfico de barras con los promedios
resumen.plot(kind='bar', figsize=(10,6))
plt.title("Promedio de variables ambientales por nivel de biomasa")
plt.ylabel("Valor promedio")
plt.xlabel("Nivel de biomasa")
plt.xticks(rotation=0)
plt.legend(title="Variable")
plt.show()
```

Este estudio nos muestra que los ecosistemas con una biomasa alta presentan una mayor disponibilidad tanto de luz, como humedad, lo que favorece su productividad, mientras que los de biomasa baja se asocian con condiciones más limitadas; los sitios de biomasa media reflejan valores intermedios, confirmando que la biomasa depende directamente de las variables ambientales y que la luz y la humedad son factores clave en la dinámica ecológica.
