## Exploración inicial
Describe las columnas y analiza los tipos de datos.

```python
import pandas as pd

# Datos de ejemplo
datos = {
    "sitio": ["A", "B", "C", "D", "E"],
    "temperatura": [18, 25, 12, 28, 15],
    "humedad": [80, 40, 90, 35, 75]
}

df = pd.DataFrame(datos)

print(df)
```

## Desarrollo del reto
Aplica funciones y análisis según el objetivo. Usa `def` para definir tus funciones.

```python
def resumen_termico(df):

    resumen = pd.DataFrame({
        "Temperatura": [
            df["temperatura"].mean(),
            df["temperatura"].max(),
            df["temperatura"].min()
        ],
        "Humedad": [
            df["humedad"].mean(),
            df["humedad"].max(),
            df["humedad"].min()
        ]
    },
    index=["Promedio", "Máximo", "Mínimo"])

    return resumen
```

```python
print("\nResumen térmico:")
print(resumen_termico(df))
```

```python
def clasificar_clima(temp, hum):

    if temp >= 20 and hum >= 60:
        return "húmedo templado"

    elif temp >= 20 and hum < 60:
        return "seco templado"

    elif temp < 20 and hum >= 60:
        return "húmedo frío"

    else:
        return "seco frío"
```

```python
df["clima"] = df.apply(
    lambda fila: clasificar_clima(
        fila["temperatura"],
        fila["humedad"]
    ),
    axis=1
)
```

```python
print("\nDataFrame con clasificación climática:")
print(df)
```

# Observaciones y conclusiones del análisis climático

## Observaciones

A partir del análisis de los datos de temperatura y humedad, se identificaron diferentes patrones climáticos asociados a cada sitio evaluado; los sitios con temperaturas iguales o superiores a 20 °C y humedad alta fueron clasificados como **húmedo templado**, indicando condiciones favorables para el desarrollo de biomasa vegetal debido a la disponibilidad de agua y temperaturas moderadas.
Los sitios con temperaturas altas y baja humedad fueron clasificados como **seco templado**, lo que puede representar zonas con mayor estrés hídrico y menor retención de humedad en el suelo. Los ambientes con temperaturas inferiores a 20 °C y alta humedad se clasificaron como **húmedo frío**, característicos de ecosistemas con menor evaporación y alta conservación de humedad.
Las zonas clasificadas como **seco frío** presentaron simultáneamente bajas temperaturas y baja humedad relativa, condiciones menos favorables para ciertos procesos biológicos y productivos. La distribución de frecuencias permitió identificar qué tipo climático predominó en el conjunto de datos analizado.
